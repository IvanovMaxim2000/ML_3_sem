## Задание 2
Написать скрипт, который меняет местами луну и голубую воронку (в самом центре) на изображении starry_night.jpg.

```python
img = cv.imread('starry_night.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

funnel = img[100:320, 280:470, :]
moon = img[40:180, 600:730, :]

funnel = cv.resize(funnel, funnel.shape[:2])
moon = cv.resize(moon, moon.shape[:2])

img[145:(145 + moon.shape[0]), 310:(310 + moon.shape[1]), :] = moon
img[20:(20 + funnel.shape[0]), 525:(525 + funnel.shape[1]), :] = funnel

plt.imshow(img)
```

## Задание 3
Попробуйте превратить "starry_night" в "starry_day" путем увеличения яркости изображения и смены темно-синих тонов на светло-желтые.

```python
img = cv.imread('starry_night.jpg',1)
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
def night_to_day(night_img, channel_koefs):
    day_img = night_img

    red_channel = day_img[:, :, 0].astype('int32')
    green_channel = day_img[:, :, 1].astype('int32')
    blue_channel = day_img[:, :, 2].astype('int32')

    day_img[:, :, 0] = red_channel / channel_koefs[0]
    day_img[:, :, 1] = green_channel / channel_koefs[1]
    day_img[:, :, 2] = blue_channel / channel_koefs[2]

    plt.imshow(cv.convertScaleAbs(night_img, day_img, 2.8, 10))


img = cv.imread('starry_night.jpg',1)
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

channel_koefs = [1.3, 1.7, 2.1]

night_to_day(img, channel_koefs)
```
