```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
from math import sin, cos, pi
```

## Задание 1
Повернуть изображение относительно его центра (а не начала координат).

```python
img = cv.imread('messi5.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
def rotation(img):
    (h, w) = img.shape[:2]
    center = (int(w / 2), int(h / 2))
    rotation_matrix = cv.getRotationMatrix2D(center, -45, 0.6)
    rotated_img = cv.warpAffine(img, rotation_matrix, (w, h))

    return rotated_img
```

```python
rotated_img = rotation(img)
plt.imshow(rotated_img)
```

## Задание 2
В папке находятся изображения (все кроме messi5.jpg), которые были подвернуты различным деформациям. Используя аффинные преобразования, необходимо "выправить" эти изображения (привести в исходный вид).

```python
img = cv.imread('apple.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
height, width = img.shape[:2]
cx, cy = width // 2, height // 2

angle = 180 * pi/ 180
M = np.float32([
    [cos(angle), -sin(angle), (1 - cos(angle)) * cx + sin(angle) * cy],
    [sin(angle), cos(angle),  (1 - cos(angle)) * cy - sin(angle) * cx]]
)

img_res = cv.warpAffine(img, M, (width, height))

plt.imshow(img_res)
```

```python
img = cv.imread('text.png')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
height, width = img.shape[:2]

warp_mat = np.float32([
    [-1, 0, width],
    [0, 1,  0]]
)

rotated_img = cv.warpAffine(img, warp_mat, (width, height))

plt.imshow(rotated_img)
```

```python
img = cv.imread('skull.png')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
height, width = img.shape[:2]
angle = -30 * pi / 180

warp_mat = np.float32([
    [0.5, -sin(angle), 0],
    [0.5 + sin(angle), 0.5, 0]
])

rotated_img = cv.warpAffine(img, warp_mat, (width, height))

plt.imshow(rotated_img)
```
