```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

## Задание 1
Используя модель HSV выполнить задание по превращению "starry_night" в "starry_day".

```python
img = cv.imread('starry_night.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
hsv_image = cv.cvtColor(img, cv.COLOR_RGB2HSV)
h,s,v = cv.split(hsv_image)
blue_mask = h > 100

hsv_image[blue_mask, 0] = 30
h,s,v = cv.split(hsv_image)
v = v + 20
hsv_image = cv.merge((h,s,v))

hsv_image = cv.cvtColor(hsv_image, cv.COLOR_HSV2RGB)
plt.imshow(hsv_image)
```

## Задание 2
Имея размытое изображение, как сделать его более четким? Попробуйте подобрать матрицу, которая может это сделать.

```python
img = cv.imread('brooklyn.webp')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
sharp_filter = np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]])
sharpen_img = cv.filter2D(img, ddepth=-1, kernel=sharp_filter)
plt.imshow(sharpen_img)
```
